# DATA 612 — Research Discussion 4: Mitigating the Harm of Recommender Systems

**Name:** Noah Collin  **Course:** DATA 612

This one picked up right where my last discussion left off, because the root problem is the same thing I was already complaining about: optimize a recommender for engagement and it learns that the way to keep you watching is to keep escalating. Zeynep Tufekci literally calls YouTube "the Great Radicalizer" in her NYT piece, and her point is that Up Next and autoplay don't have to be designed to radicalize anyone — the system just figured out, on its own, that slightly more extreme content holds attention a little longer, and "a little longer," compounded across millions of sessions, walks people from the mainstream out to the fringe. Renée DiResta's Wired article ("Up Next: A Better Recommendation System") frames it as an optimization-target problem, which is the framing that lands for me: point a very powerful optimizer at "watch time" and it will find the shortcut, and outrage happens to be a great shortcut.

The piece I genuinely hadn't thought about until I read the Krishnan, Patel, Franklin, and Goldberg paper on social influence bias is that the harm isn't only in what gets recommended — it's in the data coming back in. They show that just displaying the current average rating nudges your own rating toward it (basically anchoring). So the ratings the system collects are already contaminated by whatever it showed you, and then it trains on that contaminated data, and the herd effect quietly feeds itself. It's a sneakier loop than the radicalization one, but it's the same shape: the system is partly learning from its own past behavior instead of from honest preferences.

As for how you actually counter this stuff, the readings plus some googling pointed me at a few levers. The biggest one is just changing what you optimize for — stop maximizing raw watch time and optimize for something closer to genuine satisfaction or long-term retention, even if it costs a little short-term engagement. That's basically the diversity/novelty re-ranking I ended up building in Project 4, which was kind of satisfying to realize — trading a sliver of accuracy for a less narrowing experience is a real, implementable thing and not just a nice sentiment. Beyond that there are "circuit breakers" that notice when a session is spiraling toward more and more extreme content and break the chain; demoting borderline content so it stops getting actively pushed (without necessarily deleting it); and, on the data side, collecting a user's rating before you show them the crowd's, so the anchoring bias from the Krishnan paper doesn't creep in. And for the genuinely high-stakes stuff — ads for jobs, housing, credit — keeping a human in the loop and actually auditing who gets shown what.

What I keep coming back to is that almost none of these are algorithm problems, they're objective problems. The recommender is doing exactly what it was asked to do — the trouble is the wrong thing got asked for. It reminds me a little of optimizing some metric at work and then acting surprised when people game it — the system isn't broken, the incentive was.

I don't think there's a clean fix, and honestly the engagement incentive is so financially strong that I'm a little skeptical the big platforms give it up without outside pressure. But it was reassuring to see that the mitigations aren't science fiction — diversity injection, circuit breakers, de-biasing the feedback you collect — they're things you can build with pretty ordinary, off-the-shelf tools. The hard part isn't really technical; it's deciding that the engagement number isn't the only thing that counts.

**Sources I looked at:** Renée DiResta (2018), *Up Next: A Better Recommendation System* (Wired); Zeynep Tufekci (2018), *YouTube, the Great Radicalizer* (NYT); Krishnan, Patel, Franklin & Goldberg, *Social Influence Bias in Recommender Systems*.
