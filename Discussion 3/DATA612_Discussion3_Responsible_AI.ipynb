# DATA 612 — Research Discussion 3: Responsible AI and Recommenders

**Name:** Noah Collin  **Course:** DATA 612

The prompt here asks two things: in what ways do recommender systems reinforce human bias, and do they reinforce or actually help prevent unethical targeting and customer segmentation. My gut reaction was that they pretty much have to reinforce bias, and the reason is baked right into how they work. Whether it's content-based filtering, user-user or item-item collaborative filtering, or matrix factorization, the model learns from how people actually behaved. It has no idea whether that behavior was fair or not; it just sees a pattern and tries to reproduce it. So a recommender ends up being a kind of mirror, except a mirror that also amplifies, because the moment it shows you something and you click, that click becomes new training data that "confirms" the pattern. That feedback loop is the part that worried me the most.

The thing that made this concrete for me was a talk by Evan Estola from Meetup ("When Recommendation Systems Go Bad"). His example was that Meetup noticed fewer women were going to tech events. If you just let a recommender learn from that, it would "helpfully" start recommending fewer tech meetups to women, which means even fewer women show up, which the model then reads as more evidence that it was right. Nobody coded "discriminate against women" anywhere — the system just faithfully learned a biased world and then made it a little more biased. What stuck with me is that their fix wasn't some clever algorithm, it was a deliberate human decision to not let gender feed into those recommendations at all.

That sent me down a rabbit hole on "proxy variables," because my next thought was "okay, so just don't give the model gender." Turns out that doesn't save you. The model doesn't need a gender or race column to discriminate — it'll happily reconstruct it from other stuff that correlates, like what you've watched, where you live, or what time of day you're online. In collaborative filtering this is almost the whole mechanism: "people similar to you" very often ends up meaning "people in your same demographic," so the bias rides in through the back door even when you think you left it out front.

This is where the Hardt, Price, and Srebro paper, "Equality of Opportunity in Supervised Learning," actually changed how I was thinking. I had assumed fairness basically meant "be blind to the protected attribute," but they argue blindness doesn't work (for the proxy reason above) and instead define fairness as something you measure on the outcomes — roughly, that the model's error rates should be about the same across groups (their "equalized odds" / "equal opportunity" idea). The counterintuitive flip for me was realizing you sometimes have to look at the protected attribute to check that you're being fair, instead of pretending it isn't there. That feels backwards at first but it makes sense once the proxy thing sinks in.

So, back to the actual question — reinforce or prevent? My honest answer is that left alone they reinforce, and at a scale and speed no individual human could match, which is exactly what makes it dangerous. The same machinery that recommends me a movie also decides who sees a job ad or a housing listing, and there are real documented cases of that going sideways (the discriminatory ad-targeting stories, and the Amazon resume-screening tool that taught itself to downrank resumes containing the word "women's"). But the prevention side is real too — it's just not automatic. You have to actually measure outcomes by group, audit for those proxy leaks, and optimize for something other than raw engagement. None of that happens unless a person decides to do it on purpose.

Where my head actually went with this is that it reminds me a lot of the stakeholders-versus-shareholders argument. The real question underneath "should the recommender be fair?" is "does a company even have to think about people who aren't its customers?" — the upstream and downstream entities, the women who quietly stopped getting the tech-meetup invites, the applicants who never saw the job ad. If you decide yes, the company is responsible for those ripple effects, then you've signed up for a much more holistic re-evaluation of the whole business, not just a patch on one model. If you decide no, you optimize for the customer and the bottom line and call it a day.

And that maps almost one-to-one onto the bias question in these models. Do you want the thing to be as predictive as humanly possible, bias issues be damned, or do you want — or are you obliged — to correct it? Honestly I think the answer depends on what the consequences are. A recommender that's a little skewed about which comedy it shows me is not the same animal as one nudging who sees a housing listing or a paycheck. The stakes are what decide how hard you're actually obligated to fight the bias.

I'll also cop to the cynical half of my brain here: it's awfully convenient for outsiders to stand off to the side and announce how things "should" be done, when they're not the ones answering to investors who want the most predictive, most profitable model on the table. Sometimes ethics and good business just don't line up, and pretending they always do is its own kind of dishonesty. Correcting bias can cost you accuracy, and accuracy is frequently the exact thing being sold. That tension is real and I don't think it has a clean answer — which is maybe the most honest thing I can say about it.

If I had to sum up my own takeaway, it's that fairness isn't a property these systems have by default — it's something you have to go out of your way to build in and then keep checking, the same way you'd keep obsessively checking an error metric like RMSE on a model you actually cared about. The math itself is neutral; the choice of what to optimize and what to measure is where the ethics actually live.
